```python
satisfaction_strs = {
    'Extremely satisfied': 7,
    'Moderately satisfied': 6,
    'Slightly satisfied': 5,
    'Neither satisfied nor dissatisfied': 4,
    'Slightly dissatisfied': 3,
    'Moderately dissatisfied' : 2,
    'Extremely dissatisfied' : 1
}

stkovrflow_visited_frequency = {
    'I have never visited Stack Overflow (before today)': 0,
    'Less than once per month or monthly': 1,
    'A few times per month or weekly': 2,
    'A few times per week': 3,
    'Daily or almost daily': 4,
    'Multiple times per day': 5
}

stkovrflow_participated_frequency = {
    'I have never participated in QA on Stack Overflow': 0,
    'Less than once per month or monthly': 1,
    'A few times per month or weekly': 2,
    'A few times per week': 3,
    'Daily or almost daily': 4,
    'Multiple times per day': 5
}

```

```python
satisfy_key = ['JobSatisfaction', 'CareerSatisfaction']
stkovrflow_visited_frequency_key = ['StackOverflowVisit']
stkovrflow_participated_frequency_key = ['StackOverflowParticipate']
```

```python
replacers = [
    (satisfy_key, satisfaction_strs),
    (stkovrflow_visited_frequency_key, stkovrflow_visited_frequency),
    (stkovrflow_participated_frequency_key, stkovrflow_participated_frequency)]
```

```python
def dict_map(dict_to_use):
    def mapper(val):
        if val in dict_to_use:
            return dict_to_use[val]
        else:
            return np.nan

    return mapper
```

```python
for keys, strs in replacers:
    survey_2018[keys] = survey_2018[keys].applymap(dict_map(strs))
```

```python
survey_2018['JobSatisfaction']
```

```python
survey_2018['JobSatisfaction'].fillna(round(survey_2018['JobSatisfaction'].mean()), inplace=True)
survey_2018['CareerSatisfaction'].fillna(round(survey_2018['CareerSatisfaction'].mean()), inplace=True)
survey_2018['StackOverflowVisit'].fillna(round(survey_2018['StackOverflowVisit'].mean()), inplace=True)
survey_2018['StackOverflowParticipate'].fillna(round(survey_2018['StackOverflowParticipate'].mean()), inplace=True)
```
